$$

$$

$$
\begin{align}
\psi(x) &= \sqrt{\frac{2}{L}}\sin(\frac{n\pi}{L} x) \\
E_n = \frac{n^2\hbar^2\pi^2}{2mL^2}
\end{align}
$$

$$
\Psi(x,t) = \psi(x)e^{-iEt/\hbar}
